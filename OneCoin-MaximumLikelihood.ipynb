```julia
using Distributions
using PyPlot
```

```julia
function prob(observed_data, model_parameter)
    n1 = observed_data[1]
    n2 = observed_data[2]
    N = n1 + n2
    
    p = model_parameter[1]
    
    return pdf(Binomial(N, p), n1)
#     return binomial(N, n1) * p^n1 * (1-p)^n2
end
```

```julia
x = 0 : 0.01 : 1

observed_data = [0,1]
y= [prob(observed_data, [xi]) for xi in x];
plot(x, y, label = string(observed_data))


observed_data = [2,2]
y= [prob(observed_data, [xi]) for xi in x];
plot(x, y, label = string(observed_data))

observed_data = [1, 3]
y= [ prob(observed_data, [xi])  for xi in x];
plot(x, y, label = string(observed_data))

observed_data = [4,5]
y= [prob(observed_data, [xi]) for xi in x];
plot(x, y, label = string(observed_data))

observed_data = [91,110]
y= [prob(observed_data, [xi]) for xi in x];
plot(x, y, label = string(observed_data))

ylabel("Likelihood")
xlabel("p")
# yscale("log")
legend()
```
