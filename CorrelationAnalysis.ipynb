```julia
all = filter(c -> c in ['b', '2'], collect(all));
```

```julia
function correlation(data, dist = 1)
    n = Dict()
    for pos in 1:length(data)
        posdist = pos + dist
        if posdist > length(data)
            posdist = posdist - length(data)
        end
        k = data[pos] * data[posdist]
        
        n[k] = get(n, k, 0) + 1
    end
    n["bb"] * sum(values(n)) / ( (n["bb"] + n["2b"]) * (n["bb"] + n["b2"]))
end
```

```julia
x = 1:(length(all) รท 2)
y = map(x -> correlation(all, x), x)
scatter(x, y)
xlabel("dist")
ylabel("correlation");
```
